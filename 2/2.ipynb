```julia
# Testing out eachline
open("input.txt") do f 
    for line in eachline(f)
        println(line)
    end
end
```

```julia
# Store on each iteration
a = []
open("input.txt") do f 
    for line in eachline(f)
        append!(a, line)
    end
end

a
```
